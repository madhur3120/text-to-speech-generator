```python
import tkinter as tk
from tkinter import ttk
from tkinter import *
from tkinter import filedialog
from tkinter.ttk import Combobox
import pyttsx3
import os
from gtts import gTTS
from playsound import playsound
import PyPDF2
```

```python
# Initializing a label box
root = Tk()
root.geometry("1000x580+200+80")
root.configure(bg = 'ghost white')
root.title("TEXT TO SPEECH & DOC TO SPEECH CONVERTER")


# Conversion Code
tts = pyttsx3.init()



def speaknow():
    text = text_box.get(1.0, END)
    character = m_f_voice.get()
    speed = voice_speed.get()
    voices = tts.getProperty('voices')
    
    def setvoice():
        if(character == 'Male'):
            tts.setProperty('voice', voices[0].id)
            tts.say(text)
            tts.runAndWait()
        else:
            tts.setProperty('voice', voices[1].id)
            tts.say(text)
            tts.runAndWait()
        
    if(text):
        if(speed == 'Slow(0.5)'):
            tts.setProperty('rate', 60)
            setvoice()
        elif(speed == 'Medium(1)'):
            tts.setProperty('rate', 100)
            setvoice()
        else:
            tts.setProperty('rate', 200)
            setvoice()


def download():
    text = text_box.get(1.0, END)
    character = m_f_voice.get()
    speed = voice_speed.get()
    voices = tts.getProperty('voices')
    
    def setvoice():
        if(character == 'Male'):
            tts.setProperty('voice', voices[0].id)
            path= filedialog.askdirectory()
            os.chdir(path)
            tts.save_to_file(text, 'text.mp3')
            tts.runAndWait()
        else:
            tts.setProperty('voice', voices[1].id)
            path= filedialog.askdirectory()
            os.chdir(path)
            tts.save_to_file(text, 'text.mp3')
            tts.say(text)
            tts.runAndWait()
        
    if(text):
        if(speed == 'Slow(0.5)'):
            tts.setProperty('rate', 60)
            setvoice()
        elif(speed == 'Medium(1)'):
            tts.setProperty('rate', 100)
            setvoice()
        else:
            tts.setProperty('rate', 200)
            setvoice()


# icon image
# image_icon = PhotoImage(file="text_audio_1.png")
# root.iconphoto(False,image_icon)

# Header Label
Label(root, text = "TEXT_TO_SPEECH CONVERTER", font = "arial 20 bold", bg = "white smoke").pack()

Msg = StringVar()
Label(root, text = "Enter Text", font = 'arial 15 bold', bg = 'white smoke').place(x= 30, y=100)
# entry_field = Entry(root, textvariable = Msg , width = '50')
# entry_field.place(x = 30, y = 150, width=800 ,height=100)

# Text box to enter text
text_box = Text(root, font="calibri 30", bg="white", relief = GROOVE, wrap= WORD, bd=0)
text_box.place(x=30, y=150, width=940, height= 100)


# Voice Selector
ttk.Label(root, text = "Select Voice :", font = ("Times New Roman", 10)).place(x= 30, y=300)

n = tk.StringVar()
m_f_voice = ttk.Combobox(root, width = 27, textvariable = n)
m_f_voice.place(x= 130, y=300)
m_f_voice['values'] = ('Male','Female')
m_f_voice.set('Male')
m_f_voice.current()


# Speed Manager
ttk.Label(root, text = "Select Speed :", font = ("Times New Roman", 10)).place(x= 530, y=300)

n1 = tk.StringVar()
voice_speed = ttk.Combobox(root, width = 27, textvariable = n1)
voice_speed.place(x= 630, y=300)
voice_speed['values'] = ('Slow(0.5)','Medium(1)','Fast(2)')
voice_speed.set('Medium(1)')
voice_speed.current()


# Button
play_button = Button(root, text='Convert to audio', compound= LEFT, bg='white', width=20, font='arial 14 bold', borderwidth= '0.1c', command = speaknow)
play_button.place(x=150 ,y=350)

play_button = Button(root, text='Save', compound= LEFT, bg='white', width=20, font='arial 14 bold', borderwidth= '0.1c', command = download)
play_button.place(x=550 ,y=350)

def docx_to_speech():
    filepath = filedialog.askopenfilename(
        title="Select a DOCX file",
        filetypes=(("DOCX files", "*.docx"),)
    )
    if filepath:
        # Convert DOCX file to text
        text = docx2txt.process(filepath)
        # Speak the text
        tts.say(text)
        tts.runAndWait()

# Button
upload_button = Button(root, text="Upload DOC and Convert", compound=LEFT, bg="white", width=20, font="arial 14 bold", borderwidth='0.1c', command=docx_to_speech)
upload_button.place(x=400, y=450)


def pdf_to_speech():
    filepath = filedialog.askopenfilename(
        title="Select a PDF file",
        filetypes=(("PDF files", "*.pdf"),)
    )
    if filepath:
        with open(filepath, "rb") as file:
            # Read PDF file and create a PDF reader object
            pdf_reader = PyPDF2.PdfReader(file)
            # Extract text from PDF
            text = ""
            for page in pdf_reader.pages:
                text += page.extract_text()
            # Speak the text
            tts.say(text)
            tts.runAndWait()

# Header Label
Label(root, text="PDF_TO_SPEECH CONVERTER", font="arial 20 bold", bg="white smoke").pack()

# Button
upload_button = Button(root, text="Upload PDF and Convert", compound=LEFT, bg="white", width=20, font="arial 14 bold", borderwidth='0.1c', command=pdf_to_speech)
upload_button.place(x=400, y=500)

root.mainloop()
```
